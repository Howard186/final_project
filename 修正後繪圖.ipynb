### 導入套件及字體

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
my_font = FontProperties(fname="simhei-1.ttf")
```

### 定義繪圖函數

```python
def figure(name , figsize=(14, 8) ):
    df_1 = pd.read_csv (r'./taiwan_movie/'+ name +'.csv')
    movie =df_1[['week','totalAmounts']].reset_index().drop("index",axis=1)
    fig = plt.figure(figsize = figsize )
    plt.tick_params(axis='x', labelsize=6)
    ax = plt.axes()
    ax.plot(movie['week'].values, movie['totalAmounts'].values, label= name)
    ax.set_title('票房累計', fontproperties=my_font)
    ax.set_xlabel('週別', fontproperties=my_font)
    ax.legend(prop=my_font)
    return plt.show()
```

### 進行繪圖

```python
figure("迷失安狄")
figure("叱咤風雲")
figure("哭悲")
figure("緝魂")
figure("角頭－浪流連",figsize=(16, 8) )
figure("我沒有談的那場戀愛")
figure("跟你老婆去旅行")
figure("複身犯")
figure("當男人戀愛時")
figure("聽見歌 再唱",figsize=(18, 8) )
figure("期末考")
figure("邪佞")
figure("靈語")
figure("瀑布")
figure("青春弒戀")
figure("金錢男孩MONEYBOYS")
figure("月老")
figure("美國女孩")
figure("詭祭")
figure("詭扯")
```
